```python
import CountVectorizer
import re

train = df['description']
game_name = df['name']

def tokenizer(text):
    # Replace any non-alphanumeric characters with spaces
    text = re.sub('[^0-9a-zA-Z]+', ' ', text)
    # Split the text into words
    words = text.split()
    # Remove any words that contain digits followed by letters
    words = [word for word in words if not re.search('\d+[a-zA-Z]', word)]
    words = [word for word in words if not word.isdigit()]
    return words

# instantiate the vectorizer object
countvectorizer = CountVectorizer(analyzer= 'word', 
                                  stop_words='english',
                                  tokenizer=tokenizer)

# convert th documents into a matrix
count_wm = countvectorizer.fit_transform(train)
count_tokens = countvectorizer.get_feature_names_out()

df_countvect = pd.DataFrame(data = count_wm.toarray(),
                            index = game_name ,
                            columns = count_tokens)
```
