```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('sorted_word.csv',index_col=[0])
```

```python
droplist = []
for word in df.columns.values:
  if df[word][2000] <= 2 :
    droplist.append(word)

df = df.drop(columns=droplist)
df
```

```python
numword = df.to_numpy()[2000]
s = pd.Series(numword)
cumulative =s.cumsum()
```

```python
import matplotlib.pyplot as plt
plt.plot(cumulative)
plt.show()
```
