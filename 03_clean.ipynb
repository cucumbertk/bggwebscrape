```python
import re

def clean(text):
  text = re.sub("<.*?>", "", text)
  text = re.sub("&.*?;", "", text)
  text = re.sub("\\\\n", "", text)
  return text
```

```python
df = pd.read_csv("bgg_scraped.csv")
```

```python
df["description"] = df["description"].apply(lambda x: clean(x))
df = df[['id','rank','name', 'description']]
df
```
