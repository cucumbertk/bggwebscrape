## remove no meaning word (glove dict)


```python
import gensim
from gensim.models import KeyedVectors
from gensim.scripts.glove2word2vec import glove2word2vec

glove_input_file='glove.6B.100d.txt'
glove_output_file='word2vec.txt'

glove2word2vec(glove_input_file,glove_output_file)
model = KeyedVectors.load_word2vec_format(glove_output_file, binary = False)
```

```python
def filter_column_append(column):
  filter_list = []
  for word in column :
    if word in model.key_to_index:
      filter_list.append(word)
  return filter_list

df_filterword = df_countword[filter_column_append(column_names)]
```

```python
len(filter_column_append(column_names))
```

```python
df_filterword = df_countword[filter_column_append(column_names)]
```

## sum by frequency



```python
import pandas as pd
```

```python
df = pd.read_csv('remove_nondict.csv')
```

```python
df = df_filterword
```

```python
df.columns.values[1:]
```

```python
freqword = df.sum(axis=0)
```

```python
freqword
```

```python
freqword_without_name = freqword[1:]
freqword_without_name
```

```python
test_keys = df.columns.values
test_values = ['sum'] + list(freqword_without_name)

# using naive method
# to convert lists to dictionary
res = {}
for key in test_keys:
    for value in test_values:
        res[key] = value
        test_values.remove(value)
        break
# Printing resultant dictionary
# print("Resultant dictionary is : " + str(res))
```

```python
df_sum = df.append(res,ignore_index=True)
df_sum
```

```python
df_sum.to_csv('sumword.csv')
```
